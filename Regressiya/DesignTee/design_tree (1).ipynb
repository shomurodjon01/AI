```python
#kerekli kutubxonalarni import qilamiz.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
#data yaratib olamiz.
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Experience': [1, 3, 5, 7, 9, 11, 13, 15],
    'Salary': [300, 450, 600, 750, 900, 1050, 1200, 1350]
}
```

```python
#bu qadam yuqoridagi lug'at turidagi datani datafream ko'rinishiga o'tkazadi.
df = pd.DataFrame(data)
```

```python
#bu yerda datamizni x, y ga ajiratib olamiz.
X = df[['Age', 'Experience']]
y = df['Salary']
```

```python

#qarorlar daraxtini regressiya modelin yaratib vizual ko'rsatadi.
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
#modelni qanchalik yaxshi o'rganganini belgilab olamiz
tree = DecisionTreeRegressor(max_depth=1)
```

```python
#modelga o'rgatamiz.
tree.fit(X, y)
```

```python
#bashorat uchun yangi malumot yaratdim.
new= pd.DataFrame({'Age': [20],'Experience': [1] })
```

```python
#datani bashorat qil degan buyruq.
predicted_sales = tree.predict(new)
```

```python
#boshoratni ko'rish.
print("Natija:", predicted_sales[0])
```

```python
#grafig chizma o'lchami.
plt.figure(figsize=(15, 12))
#daraxni chizma ko'rinishda chiqradi.
plot_tree(tree, feature_names=X.columns, filled=True, rounded=True)
plt.title("Decision Tree ")
plt.show()
```
