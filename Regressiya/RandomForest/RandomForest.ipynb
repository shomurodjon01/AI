```python
#kerakli kutubxonalar
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
data = {
    'Age': [20, 19, 25, 28, 32, 25, 29, 38, 29, 19],
    'internship': [1, 5, 8, 2, 11, 2, 9, 13, 4, 2],
    'salary': [2000, 8000, 1200, 1500, 1100, 2500, 7010, 1000, 2000, 1600],
    'class': ['1-toifa', 'oliy_toifa', 'oliy_toifa', '1-toifa', '1-toifa', '1-toifa', 'oliy_toifa', '1-toifa', 'oliy_toifa', 'oliy_toifa']
}

ish = pd.DataFrame(data)
```

```python
sns.countplot(data=ish, x='class')
plt.title("Sinf taqsimoti")
plt.xlabel("Sinf")
plt.ylabel("Odamlar soni")
plt.show()
sns.scatterplot(data=ish, x='Age', y='salary', hue = "class", style= "class", s = 100)
plt.title("Yosh va Maosh boâ€˜yicha classlar farqlari")
plt.show()
```

```python
X = ish[['Age', 'internship', 'salary']]
y = ish['class']
```

```python
y = y.map({'1-toifa': 0, 'oliy_toifa': 1})
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
comparison_ish = pd.DataFrame({'aniq_qiymat': y_test.values, 'Bashorat': y_pred})
comparison_ish.value_counts().unstack().plot(kind='bar', rot=6)
plt.title("Aniq va Bashorat qilingan qiymatlar")
plt.xlabel("class")
plt.ylabel("ishchilar soni")
plt.show()
```

```python
acc = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print("Aniqlik:", acc)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
